<h1> Confirmed Covid-19 cases evolution </h1>
This notebook visualizes the evolution of confirmed Covid-19 cases in China, US, UK, Italy, France, Germany, Spain and Iran from Jan 22nd 2020 until May 28th 2020. To visualize the change in number of cases dynamically I am using the <i> bar chart race </i> package.

```python
import pandas as pd
import bar_chart_race as bcr
from IPython.display import HTML
```

First let's read the data

```python
df = pd.read_csv('key-countries-pivoted.csv', index_col = 0)
#df.loc[:,'Date'] = pd.to_datetime(df.loc[:,'Date'])
df.info()
```

Create a method that summarizes the total number of deaths

```python
def summary(values, ranks):
    total_cases = int(round(values.sum(), -2))
    s = f'Total Cases : {total_cases:,.0f}'
    return {'x': .99, 'y': .05, 's': s, 'ha': 'right', 'size': 8}
```

Produce the chart using the <i> bar chart race </i> package.

```python
bcr_html = bcr.bar_chart_race(df, title='COVID-19 Cases by Country', label_bars=True, period_summary_func=summary)
HTML(bcr_html)
```
